```python
import numpy as np
import matplotlib.pyplot as plt
import control as ctrl

# Initialize lists to store poles and zeros
poles_list = []
zeros_list = []

# Vary the gain K from 0 to 50
K_values = np.arange(0, 50.1, 0.5)

for K in K_values:
    # Define the transfer function for the current K
    num = [K]
    den = [1, 6, -16 + K]

    # Create the transfer function
    system = ctrl.TransferFunction(num, den)

    # Calculate the poles and zeros of the transfer function
    poles = ctrl.pole(system)
    zeros = ctrl.zero(system)

    poles_list.append(poles)
    zeros_list.append(zeros)

# Plot the Zero-Pole Diagram
plt.figure()
for poles, zeros in zip(poles_list, zeros_list):
    plt.scatter(np.real(zeros), np.imag(zeros), marker='o', color='b')
    plt.scatter(np.real(poles), np.imag(poles), marker='x', color='r')

plt.title("Zero-Pole Diagram for Varying K")
plt.xlabel('Real Part')
plt.ylabel('Imaginary Part')
plt.grid(True)
plt.legend(['Zeros', 'Polos'])
plt.show()
```

```python
import numpy as np
from sympy import symbols, Eq, solve

# Defina a variável simbólica 's'
s = symbols('s')

# Defina o denominador da função de transferência
denominator = s * (s + 1) * (s**2 + 4*s + 16)

# Encontre as raízes do denominador
roots = solve(Eq(denominator, 0), s)
roots = [complex(root) for root in roots]  # Convert as complex numbers

print("Raízes da malha aberta (denominador):", roots)
```

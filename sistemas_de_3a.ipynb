```python
import numpy as np
import matplotlib.pyplot as plt
import control as ctrl

a = 3
k1 = 5/a
# Denominador da função de transferência
den = [1, 2, 5]

# Numerador da função de transferência (K * (s + a))
num = [k1, k1*a]

# Crie a função de transferência
G = ctrl.TransferFunction(num, den)

# Tempo de amostragem
t = np.linspace(0, 10, 1000)

# Calcule a resposta ao degrau
time, response = ctrl.step_response(G, T=t)

# Plote a resposta ao degrau
plt.plot(time, response)
plt.title('Resposta ao degrau de G(s) com a = {a}')
plt.xlabel('Tempo')
plt.ylabel('Resposta ao degrau')
plt.grid()
plt.show()
```

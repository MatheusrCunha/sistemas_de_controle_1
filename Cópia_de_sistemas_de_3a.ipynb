```python
import numpy as np
import matplotlib.pyplot as plt
import control as ctrl


z = [3, 5, 10]

# Tempo de amostragem
t = np.linspace(0, 10, 1000)

plt.figure()
for a in z:
    K = 5 / a

    # Denominador da função de transferência
    den = [1, 2, 5]

    num = [K, K * a]

    # Crie a função de transferência
    G = ctrl.TransferFunction(num, den)

    # Calcule a resposta ao degrau
    time, response = ctrl.step_response(G, T=t)

    # Plote a resposta ao degrau
    plt.plot(time, response, label=f'a = {a}')

plt.title('Respostas ao degrau de G(s) com diferentes valores de a')
plt.xlabel('Tempo')
plt.ylabel('Resposta ao degrau')
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import control as ctrl


z = [-1, -3, -5]

# Tempo de amostragem
t = np.linspace(0, 10, 1000)

plt.figure()
for a in z:

    den = [1, 2, 5]

    num = [-1, -1 * a]

    G = ctrl.TransferFunction(num, den)

    time, response = ctrl.step_response(G, T=t)

    plt.plot(time, response, label=f'a = {a}')

plt.title('Respostas ao degrau de G(s) = -1 * (s + a) / (s^2 + 2s + 5) para diferentes valores de a')
plt.xlabel('Tempo')
plt.ylabel('Resposta ao degrau')
plt.grid()
plt.legend()
plt.show()
```

O tempo de estabilização muda quando você altera o valor de a.

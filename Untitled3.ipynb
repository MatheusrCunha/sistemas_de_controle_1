```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import lti, step

# Defina a função de transferência do sistema com o ganho ajustado K
numerator = [-7]
denominator = [1, 2, 9]
system = lti(numerator, denominator)

# Tempo de amostragem
t = np.linspace(0, 10, num=1000)

# Calcule a resposta ao degrau
t, y = step(system, T=t)

# Plote a resposta ao degrau
plt.plot(t, y)
plt.xlabel('Tempo')
plt.ylabel('Resposta ao Degrau')
plt.title('Resposta ao Degrau do Sistema')
plt.grid(True)
plt.show()
```
